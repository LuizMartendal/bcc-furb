03 â€“ Abrir uma imagem colorida em RGB, visualizar e salvar cada um dos canais
separadamente. Obs: Busquem compreender o que significa cada um dos canais.

```
import cv2
import numpy as np
from matplotlib import pyplot as plt

imagem = cv2.imread('LenaC.jpg', cv2.IMREAD_COLOR)

imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

# Separar os canais RGB
R, G, B = cv2.split(imagem_rgb)

# Criar imagens com cada canal individual (mantendo o formato RGB)
zeros = np.zeros_like(R)

imagem_R = cv2.merge([R, zeros, zeros])  # Apenas Vermelho
imagem_G = cv2.merge([zeros, G, zeros])  # Apenas Verde
imagem_B = cv2.merge([zeros, zeros, B])  # Apenas Azul

plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(imagem_R)
plt.title('Canal Vermelho (R)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(imagem_G)
plt.title('Canal Verde (G)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(imagem_B)
plt.title('Canal Azul (B)')
plt.axis('off')

plt.show()

# Salvar os canais separadamente
cv2.imwrite('canal_R.jpg', cv2.cvtColor(imagem_R, cv2.COLOR_RGB2BGR))
cv2.imwrite('canal_G.jpg', cv2.cvtColor(imagem_G, cv2.COLOR_RGB2BGR))
cv2.imwrite('canal_B.jpg', cv2.cvtColor(imagem_B, cv2.COLOR_RGB2BGR))
```

04 â€“ Abrir uma imagem colorida, transformar em tom de cinza, visualizar imagem de
entrada. Apliquem os filtros passa baixa mediana (median) e media (blur), visualizem os resultados
e salvem. Obs: busquem compreender os resultados de cada filtro.

```
import cv2
import numpy as np
from matplotlib import pyplot as plt

imagem = cv2.imread('LenaC.jpg', cv2.IMREAD_COLOR)

imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
imagem_mediana = cv2.medianBlur(imagem_cinza, 5)
imagem_media = cv2.blur(imagem_cinza, (5, 5))

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Tons de Cinza')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagem_mediana, cmap='gray')
plt.title('Filtro Mediana')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(imagem_media, cmap='gray')
plt.title('Filtro MÃ©dia')
plt.axis('off')

plt.show()

cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
cv2.imwrite('imagem_mediana.jpg', imagem_mediana)
cv2.imwrite('imagem_media.jpg', imagem_media)
```

05 â€“ Abrir uma imagem colorida, transformar em tom de cinza, visualizar imagem de
entrada. Apliquem uma limiarizaÃ§Ã£o (thresholding), visualizem os resultados e salvem. Obs: busquem
compreender os resultados da tÃ©cnica.

```
import cv2
import numpy as np
from matplotlib import pyplot as plt

imagem = cv2.imread('LenaC.jpg', cv2.IMREAD_COLOR)
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

_, imagem_threshold = cv2.threshold(imagem_cinza, 127, 256, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Tons de Cinza')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagem_threshold, cmap='gray')
plt.title('LimiarizaÃ§Ã£o (Thresholding)')
plt.axis('off')

plt.show()

cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
cv2.imwrite('imagem_threshold.jpg', imagem_threshold)
```

06 â€“ Abrir uma imagem colorida, transformar para tom de cinza e aplicar uma EqualizaÃ§Ã£o
de histograma utilizando a OpenCv, visualizando a imagem de entrada e seu respectivo histograma
inicialmente, e, em seguida, o resultado da equalizaÃ§Ã£o e seu histograma. Esta tÃ©cnica aumenta o
contraste da imagem

```
import cv2
import numpy as np
from matplotlib import pyplot as plt

imagem = cv2.imread('LenaC.jpg', cv2.IMREAD_COLOR)

imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
hist_original = cv2.calcHist([imagem_cinza], [0], None, [256], [0, 256])

imagem_equalizada = cv2.equalizeHist(imagem_cinza)
hist_equalizado = cv2.calcHist([imagem_equalizada], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem Original em Tons de Cinza')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.plot(hist_original)
plt.title('Histograma da Imagem Original')
plt.xlabel('Intensidade de Pixel')
plt.ylabel('NÃºmero de Pixels')
plt.xlim([0, 256])

plt.subplot(2, 2, 3)
plt.imshow(imagem_equalizada, cmap='gray')
plt.title('Imagem Equalizada')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.plot(hist_equalizado)
plt.title('Histograma da Imagem Equalizada')
plt.xlabel('Intensidade de Pixel')
plt.ylabel('NÃºmero de Pixels')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()

cv2.imwrite('imagem_equalizada.jpg', imagem_equalizada)
```
